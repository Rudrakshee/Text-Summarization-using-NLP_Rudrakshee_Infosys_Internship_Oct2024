```python
# Load text file and inspect the data
with open("data.txt", "r") as file:
    text = file.read()

# Print the first 500 characters to check the content
print(text[:500])
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk import pos_tag
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet

# Download required NLTK datasets
nltk.download("punkt")
nltk.download("stopwords")
nltk.download("wordnet")
nltk.download("omw-1.4")
```

```python

# Step 1: Convert text to lowercase
text_lower = text.lower()
```

```python

# Step 2: Tokenize the text into words
tokens = word_tokenize(text_lower)
```

```python

# Step 3: Remove stopwords
stop_words = set(stopwords.words("english"))
tokens_cleaned = [word for word in tokens if word not in stop_words and word.isalpha()]
```

```python

# Step 4: Lemmatization with POS tagging
lemmatizer = WordNetLemmatizer()
```

```python




# Helper function to map POS tags to WordNet POS tags
def get_wordnet_pos(treebank_tag):
    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN

```

```python

# Lemmatize words with POS tagging
tokens_lemmatized = [
    lemmatizer.lemmatize(word, get_wordnet_pos(tag)) for word, tag in pos_tag(tokens_cleaned)
]

# Print cleaned and lemmatized tokens
print(tokens_lemmatized[:20])
```

```python
from transformers import pipeline

# Load the BERT-based summarizer pipeline
summarizer = pipeline("summarization")

# Function to summarize the text
def summarize_text(text):
    # Summarize the cleaned text using BERT
    summary = summarizer(text, max_length=150, min_length=10, do_sample=False)
    return summary[0]['summary_text']

# Summarize the text
summary = summarize_text(text)
print("Summary:", summary)
```

```python
import gradio as gr
from transformers import pipeline

# Specify the model explicitly
model_name = "sshleifer/distilbart-cnn-12-6"
summarizer = pipeline("summarization", model=model_name)

# Function to summarize the input text
def summarize_text(text):
    if not text.strip():
        return "Please enter some text to summarize."

    # Summarize the input text
    try:
        summary = summarizer(text, max_length=200, min_length=50, do_sample=False)
        return summary[0]['summary_text']
    except Exception as e:
        return f"Error during summarization: {e}"

# Custom CSS to enhance the UI appearance
css = """
body {
    background-color: #f0f8ff; /* Light blue background */
    font-family: 'Arial', sans-serif; /* Font for the whole page */
}

#component-1 {
    background-color: #ffffff; /* White background for the input box */
    border: 2px solid #0078d4; /* Blue border for the input box */
    border-radius: 8px;
    padding: 10px;
    box-shadow: 0 4px 8px rgba(0, 0, 0, 0.1);
    font-size: 16px;
    color: #333333;
}

#component-2 {
    background-color: #0078d4; /* Blue background for the summary box */
    color: white;
    border-radius: 8px;
    padding: 10px;
    font-size: 16px;
}

h1, .gradio-interface {
    color: #2e3d49; /* Dark color for the title */
    text-align: center;
    font-family: 'Verdana', sans-serif;
}

.gradio-interface .output-box {
    background-color: #0078d4; /* Matching blue color for output box */
    color: white;
    font-size: 16px;
    padding: 15px;
    border-radius: 8px;
}

.gradio-interface .input-box {
    background-color: #ffffff;
    color: #333333;
    font-size: 16px;
    border-radius: 8px;
    padding: 15px;
}

"""

# Create the Gradio interface
interface = gr.Interface(
    fn=summarize_text,
    inputs=gr.Textbox(label="Enter Text Here", lines=10, placeholder="Type your text to summarize..."),  # Text input field
    outputs=gr.Textbox(label="Summary"),  # Output as a textbox
    live=True,  # Update the summary in real-time if desired
    title="Text Summarization",
    description="Enter text into the input box and get a concise summary of its content.",
    css=css  # Apply the custom CSS
)

# Launch the Gradio interface
interface.launch()
```
